```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
plt.plot([1, 2, 3], [2, 5, 7])
plt.show()
```

```python
plt.plot([1, 2, 3], [2, 5, 7])
plt.title("Sample Data")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.show()
```

```python
x1 = [1, 5, 4]
y1 = [23, 45, 89]
x2 = [3, 7, 3]
y2 = [43, 56, 64]

plt.plot(x1, y1, label="first", linewidth=3)
plt.plot(x2, y2, label="second")
plt.legend()
plt.show()
```

```python
x1 = [1, 5, 4]
y1 = [23, 45, 89]
x2 = [3, 7, 3]
y2 = [43, 56, 64]
style.use('ggplot')
plt.plot(x1, y1, label="first", linewidth=3)
plt.plot(x2, y2, label="second")
plt.legend()
plt.show()
```

```python
x1 = [1, 5, 4]
y1 = [23, 45, 89]
x2 = [3, 7, 3]
y2 = [43, 56, 64]
style.use('ggplot')
plt.scatter(x1, y1, label="first", linewidth=3)
plt.scatter(x2, y2, label="second")
plt.legend()
plt.show()
```
